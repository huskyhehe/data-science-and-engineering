```python
'''The usual...package imports...'''
import urllib.request
import re
from bs4 import BeautifulSoup

my_page = urllib.request.urlopen("http://www.microstrategy.com/us/company/contact-us")

# An example is "http://www.microstrategy.com/us/company/contact-us"
my_page_text = str(my_page.read())
```

```python
'''Dare to even take a look at what just came in ?Ulp !'''
print(my_page_text)
```

```python
'''Time to find them emails...using regular expressions !
Regular expression = many, many cups of coffee, tea or well, whatever
you drink...'''
email_list = re.findall(">(\w*\S*\w*@\w*\.\w+)",my_page_text)

print(email_list)
```

```python
'''We changed our mind and now we want all the phone numbers from
the page. How do we go about it ?
1. Find the pattern we are searching for and,
2. Use findall

First, could we make the mess a little less messy ?'''

soup = BeautifulSoup(my_page_text, 'html.parser')
print(soup.prettify())
```

```python
'''Time to find them phone numbers this time...
using regular expressions !Let's put on our thiking caps...
From the prettified mess above, what do we see ?
1. The phone number is a string of numbers
2. They are groupedinto sets of three
3. There's a non-word character, the hypen, between each group and,
4. The entire number itself is between > and < 
5. Use the cheat sheet (https://www.debuggex.com/cheatsheet/regex/python)
'''

telnumber_list = re.search(">\d+<",my_page_text)

print(telnumber_list)
```

```python
'''Let's try again'''

telnumber_list = re.search("tel:(\d+)",my_page_text)

print(telnumber_list)
'''oops, we only got the first three numbers from a number!'''
```

```python
'''And again, this time looking out for the hypens too
\d for digit, i.e a number and \W for non-word character'''

telnumber_list = re.search("tel:(\d+\W)",my_page_text)

print(telnumber_list)
```

```python
'''We are gettign there. There's more numbers after the hypen!
So do look for \d+ again !'''

telnumber_list = re.search("tel:(\d+\W\d+)",my_page_text)

print(telnumber_list)
```

```python
'''Not quite. Repeat !'''
telnumber_list = re.search("tel:(\d+\W\d+\W\d+)",my_page_text)

print(telnumber_list)
```

```python
'''Yaaay ! But, Ram, go to the webpage and you will see
there are three phone numbers at least ! So, let's stop using
re.search and instead use our old friend, re.findall.
Notice the difference in results ?'''

telnumber_list = re.findall("tel:(\d+\W\d+\W\d+)",my_page_text)

print(telnumber_list)
```
